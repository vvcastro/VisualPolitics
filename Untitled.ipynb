```python
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
# load dataset
with open('diputades_data.json', 'r') as dip_index:
    dip_data = json.load(dip_index)

pd_data = {}
for key, val in dip_data.items():
    new_val = { k: v for k,v in val.items() if (k != 'data') }
    for k in val['data']:
        new_val[k] = val['data'][k]
    pd_data[key] = new_val
    
dip_data = pd.DataFrame(pd_data).transpose()
```

```python
dip_data.sort_values(by="TRASLACIÓN", ascending=False).head(n=30)
```

```python
gastos_partidos = {}
for party, party_data in dip_data.groupby("party"):
    gastos = party_data[ [col for col in party_data.columns if (col not in ['name', 'district', 'party', 'image'])] ]
    gastos_partidos[party] = gastos.sum().sum() / len( party_data ) 

gastos_partidos = pd.Series(gastos_partidos).reset_index(drop=False).rename({"index": "Partido", 0: "Gasto [Millones]" }, axis=1)
gastos_partidos["Gasto [Millones]"] = gastos_partidos["Gasto [Millones]"] / 1e6
gastos_partidos.sort_values(by="Gasto [Millones]").describe()
```
